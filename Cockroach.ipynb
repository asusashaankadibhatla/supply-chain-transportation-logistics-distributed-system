```python
import subprocess
import time

def start_cockroach_node(node_id, http_port, store_dir, join_addrs):
    cmd = [
        "cockroach", "start",
        "--insecure",
        f"--store={store_dir}",
        f"--listen-addr=localhost:{26256 + node_id}",
        f"--http-addr=localhost:{http_port}",
        f"--join={','.join(join_addrs)}",
        "--background"
    ]
    subprocess.Popen(cmd, shell=True)
    subprocess.run(cmd)

def initialize_first_node():
    cmd = [
        "cockroach", "init",
        "--insecure",
        f"--host=localhost:26257"
    ]
    subprocess.Popen(cmd, shell=True)
    subprocess.run(cmd)

def get_cockroach_processes():
    result = subprocess.run(["ps", "-ef"], stdout=subprocess.PIPE)
    cockroach_processes = [line.decode() for line in result.stdout.splitlines() if b"cockroach" in line and b"grep" not in line]
    return cockroach_processes

def kill_cockroach_processes():
    cockroach_processes = get_cockroach_processes()
    pids = [line.split()[1] for line in cockroach_processes]
    pids.sort(reverse = True)
    for pid in pids:
        subprocess.run(["kill", "-9", pid])

if __name__ == "__main__":
    num_nodes = 10
    base_port = 8080

    start_cockroach_node(1, base_port, "node1", [f"localhost:{26256 + j}" for j in range(1,num_nodes+1)])

    time.sleep(5)

    initialize_first_node()

    for i in range(2, num_nodes + 1):
        start_cockroach_node(i, base_port + i - 1, f"node{i}", [f"localhost:{26256 + j}" for j in range(1,num_nodes+1)])

    print("CockroachDB nodes started and initialized successfully.")
```

```python
kill_cockroach_processes()
print("CockroachDB nodes deleted successfully.")
```
